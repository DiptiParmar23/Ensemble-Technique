```python
1756/3756*100
```

Note: There is imbalance in Target variable with almost 54% data in 0 and 46% data in 1. we can fix it by using
      undersampling method. This can be done by using sample option in pandas.

```python
sample_data=pd.concat([tc23[tc23['Churn']==0].sample(1869),tc23[tc23['Churn']==1]])
X=sample_data.drop('Churn',axis=1)
y=sample_data['Churn']
```

```python
#It is balanced data
y.value_counts()
```

### 5. Model training, testing and tuning: 

### • Train and test all ensemble models taught in the learning module.

### • Display the classification accuracies for train and test data.

#### 1. Decision Tree with entropy(model1)

```python
from sklearn.tree import DecisionTreeClassifier
dt_model1=DecisionTreeClassifier(criterion='entropy')
dt_model1.fit(X_train,y_train)
test_pred1=dt_model1.predict(X_test)

print(dt_model1.score(X_train,y_train))
#print(dt_model1.score(X_test,y_test))
print(accuracy_score(y_test,test_pred1))
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(classification_report(y_test,test_pred1))
```

### 2. Decision Tree with gini(model2)

```python
from sklearn.tree import DecisionTreeClassifier
dt_model2=DecisionTreeClassifier(criterion='gini',random_state=1)
dt_model2.fit(X_train,y_train)
test_pred2=dt_model2.predict(X_test)

print(dt_model2.score(X_train,y_train))
#print(dt_model2.score(X_test,y_test))
print(accuracy_score(y_test,test_pred2))
```

```python
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
print(classification_report(y_test,test_pred2))
```

Here the accuracy score for the train data is high which is around 100%, where in Test data is low around 66%. which means
the model is overfitted on Training data. this is expected in terms of Tree based modelling as it tend to overfit on the 
Training data. so let us try to use various techniques to find which works best for this dataset.

### 3. Ensemble technique using weak classifiers

### 3a. RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
rfcl=RandomForestClassifier(criterion='entropy',class_weight={0:.5,1:.5},max_depth=5,min_samples_leaf=5)
rfcl=rfcl.fit(X_train,y_train)
test_pred3=rfcl.predict(X_test)
#test_pred2=dt_model2.predict(X_test)
print(rfcl.score(X_train,y_train))
print(accuracy_score(y_test,test_pred3))
print(classification_report(y_test,test_pred3))
```

### AdaBoostClassifier

```python
from sklearn.ensemble import AdaBoostClassifier
abcl=AdaBoostClassifier(n_estimators=20)
abcl=abcl.fit(X_train,y_train)
test_pred4=abcl.predict(X_test)
print(abcl.score(X_train,y_train))
print(accuracy_score(y_test,test_pred4))
print(classification_report(y_test,test_pred4))
```

### AdaBoostClassifier with best estimator

```python
from sklearn.ensemble import AdaBoostClassifier
abcl=AdaBoostClassifier(base_estimator=dt_model1,n_estimators=50)
abcl=abcl.fit(X_train,y_train)
test_pred5=abcl.predict(X_test)
print(abcl.score(X_train,y_train))
print(accuracy_score(y_test,test_pred5))
print(classification_report(y_test,test_pred5))
```

### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
gbcl=GradientBoostingClassifier(n_estimators=50,learning_rate=0.05)
gbcl=gbcl.fit(X_train,y_train)
test_pred6=gbcl.predict(X_test)
print(gbcl.score(X_train,y_train))
print(accuracy_score(y_test,test_pred6))
print(classification_report(y_test,test_pred6))
```

### Bagging

```python
from sklearn.ensemble import BaggingClassifier
bgcl=BaggingClassifier(n_estimators=10,max_samples=.7,bootstrap=True)
bgcl=bgcl.fit(X_train,y_train)
test_pred7=bgcl.predict(X_test)
print(bgcl.score(X_train,y_train))
print(accuracy_score(y_test,test_pred7))
print(classification_report(y_test,test_pred7))
```

#### Conclusion: GradientBoostingClassifier seems to work better compared to other models. Bias and variance of the model seems to be similar. Other 3 models are seems to be overfitting. So, lets take GradientBoostingClassifier and fine tune with Hyperparameters. Also, the recall and precision values are more or less similar. f1 score is around 75% which tells us the models works find on the dataset. Further fine tunning might help to improve the model accuracy.
